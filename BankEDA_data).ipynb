```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar', color='green')
plt.title('Income Band Distribution')
plt.xlabel('Income Band')
plt.ylabel('Count')
plt.show()
```

```python
# Examine the distribution of unique cataegories in categorical columns
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for col in categorical_cols:
  print(f"Value Counts for '{col}':")
  display(df[col].value_counts())
```

## Univariate Analysis

```python
# Univariate Analysis
# Apply a beautiful color combination to the countplot
categorical_cols = ["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]

for i, predictor in enumerate(df[categorical_cols].columns):
    plt.figure(i)
    # Using the 'viridis' palette for a beautiful color combination
    sns.countplot(data=df, x=predictor, palette='viridis')
    plt.title(f'Countplot of {predictor}')
    plt.xlabel(predictor)
    plt.ylabel('Count')
    plt.show()
```

## Bivariate Analysis

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
# HIstplot of value counts for different Occupation

# Get a colorful palette, for example 'viridis'
color_palette = sns.color_palette('viridis')
num_colors = len(color_palette)

for i, col in enumerate(categorical_cols):
  if col == "Occupation":
    continue
  plt.figure(figsize=(6,3))
  # Use a color from the palette, cycling through it
  sns.histplot(df[col], color=color_palette[i % num_colors])
  plt.title(f'Histogram of {col} Count') # Modified title to be more accurate
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

## Numerical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(15,10))
# Get a colorful palette
color_palette = sns.color_palette('viridis')
num_colors = len(color_palette)

for i,col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  # Use a color from the palette, cycling through it
  sns.histplot(df[col],kde=True, color=color_palette[i % num_colors])
  plt.title(col)

# Use plt.tight_layout() to prevent titles/labels from overlapping
plt.tight_layout()
plt.show()
```

## Heatmaps

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## Insights of EDA:

1. The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.
